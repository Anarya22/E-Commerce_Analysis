```python
df['Product Base Margin'].fillna(df['Product Base Margin'].mean(), inplace = True)
```

```python
df['Order Priority'].value_counts()
```

```python
df['Order Priority'].unique()
```

```python
df['Order Priority'].replace("Critical ", "Critical", inplace=True)
df['Order Priority'].value_counts()
```

```python
plt.figure(figsize=[5,4])
sns.countplot(x='Order Priority', data=df)
plt.title("Count of Order Priority")
plt.savefig("Count-of-Order-Priority.jpg");
```

# Ship Mode(Univariate Analysis)

```python
df.head(2)
```

```python
df['Ship Mode'].value_counts()
```

```python
labels = list(df['Ship Mode'].value_counts().index)
sizes = list(df['Ship Mode'].value_counts().values)
labels, sizes
```

```python
plt.figure(figsize=[5,4])
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=[0.03]*len(sizes), startangle=30)
plt.legend(loc="upper right");
```
